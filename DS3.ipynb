```python
#Exercise1
def prime(num) :
    if num <=1 :
       print("Not a prime number.")
       return
    for i in range(2,int(num**0.5)+1) :
        if num%i==0 :
            print("Not a prime number.")
            return
    print("It is a prime number.")

prime(55)
```

```python
#Exercise2
def product_of_numbers() :
    a = int(input("Enter the first number :"))
    b = int(input("Enter the second number :"))
    p = int(input("Enter the product :"))

    if p == (a*b) :
        print("The calculation is correct")
    else :
        print("The calculation is incorrect")

product_of_numbers()
```

```python
#Exercise3
def even_odd() :
    choice = input("Enter a choice (even/odd) :")
    
    i=0
    for i in range(100,201) :
        if choice=="even" and i%2==0 :
            print(f"The square of all even numbers is : {i**2}")
        elif choice=="odd" and i%2!=0 :
            print(f"The square of all odd numbers is : {i**2}")

even_odd()
```

```python
#Exercise4
input_text = "This is a sample text . This text will be used to demonstrate the word counter."

words = input_text.split()
word_count={}

for word in words :
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

for word,count in word_count.items():
    print(f"{word} : {count}")
```

```python
#Exercise5
def palindrome() :
    text = input("Enter a text, word or phrase :").strip().lower()
    if text == text[::-1] :
        print("It is a palindrome !")
    else:
        print("It is not a palindrome.")

palindrome()
```
