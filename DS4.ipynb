```python
#Task1
# H0 (Null Hypothesis): Customer satisfaction level is independent of the type of device purchased
# H1 (Alternative Hypothesis): Customer satisfaction level is associated with the type of device purchased
```

```python
#Task2&3&4
import scipy.stats as stats
import numpy as np
```

```python
observed = np.array([
    [50,70],
    [80,100],
    [60,90],
    [30,50],
    [20,50]
])

chi2_stat,p_value,dof,expected = stats.chi2_contingency(observed)
print("Chi-Square Test for Independence")
print("--------------------------------")
print("Observed Frequencies:\n", observed)
print("\nExpected Frequencies:\n", np.round(expected, 2))
print("\nChi-Square Statistic:", round(chi2_stat, 4))
print("Degrees of Freedom:", dof)
print("P-value:", round(p_value, 4))

alpha = 0.05
critical_value = stats.chi2.ppf(1-alpha,dof)
print("Critical Value at alpha = 0.05 :",round(critical_value,4))
```

```python
if chi2_stat > critical_value :
    print("\nDecision : Null Hypothesis Rejected .")
    print("There is association between the type of device purchased and customer satisfaction")
else :
    print("\nDecision : Failed to reject Null Hypothesis rejected.")
    print("Both the variables are independent.")
```

```python
# Hypothesis Testing Assignment
# Bombay Hospitality Ltd.

import numpy as np
from scipy.stats import norm

# -------------------------------
# 1. Hypothesis
# H0: μ = 4000 (weekly cost follows the model W = 1000 + 5X)
# H1: μ > 4000 (weekly costs are higher than predicted)
# -------------------------------

# Given values
x_bar = 3050                # sample mean weekly cost
X_mean = 600                # mean units
sigma_X = 25                # SD of units produced
sigma_W = 5 * sigma_X       # SD of weekly cost
n = 25                      # sample size

# Theoretical weekly cost using the model
mu_0 = 1000 + 5 * X_mean

# -------------------------------
# 2. Test Statistic (Z-test)
# -------------------------------

SE = sigma_W / np.sqrt(n)   # standard error
Z = (x_bar - mu_0) / SE      # test statistic

print("Hypothesis Testing for Weekly Operating Costs\n")
print("Theoretical mean cost (μ0):", mu_0)
print("Sample mean (x̄):", x_bar)
print("Standard deviation of cost (σ):", sigma_W)
print("Standard Error (SE):", SE)
print("Test Statistic (Z):", Z)

# -------------------------------
# 3. Critical Value
# Right-tailed test at alpha = 0.05
# -------------------------------

alpha = 0.05
Z_critical = norm.ppf(1 - alpha)

print("\nCritical Value (Z_critical):", Z_critical)

# -------------------------------
# 4. Decision
# -------------------------------

if Z > Z_critical:
    print("\nDecision: Reject H0")
else:
    print("\nDecision: Fail to Reject H0")

# -------------------------------
# 5. Conclusion
# -------------------------------

print("\nConclusion:")
if Z > Z_critical:
    print("There is strong evidence that weekly operating costs are higher than the model suggests.")
else:
    print("There is no statistical evidence that weekly operating costs are higher.")
    print("The model's cost prediction appears valid.")
```
