```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing dataset
df = pd.read_csv("Cardiotocographic.csv")
```

```python
#Data Preprocessing Steps

print("\nBasic Info : ")
print(df.info())
print("\nStatistical Summary : ")
print(df.describe())
print("\nShape of data : ")
print(df.shape)
print("\nMissing null values per column:",df.isnull().sum())
df.fillna(df.median(), inplace=True)
print("\nInitial Values :")
print(df.head())
```

```python
#Visualizations

df.hist(figsize=(10,8),bins=20,color="skyblue",edgecolor="black")
plt.title("Histogram of Numeric Features")
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.title("Boxplots of Features")
plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(),annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

cols = ["LB","AC","FM","UC","ASTV","MSTV"]
sns.pairplot(df[cols])
plt.title("Pairplots for Selected Variables")
plt.show()

print("\n EDA Completed Successfully !")
print("--------Observations--------")
print("1.Check Boxplots for possible outliers.")
print("2.Use Heatmap to find correlation between variables.")
```
